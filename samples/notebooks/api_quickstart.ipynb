# histdatacom API quickstart

This notebook mirrors the README API sample. Running it normally uses the real histdatacom API. The test suite sets `HISTDATACOM_SAMPLE_MODE=hermetic` so the same notebook executes without live services.

```python
import os

sample_context = None
if os.environ.get("HISTDATACOM_SAMPLE_MODE") == "hermetic":
    from samples._testing import install_fake_orchestration

    sample_context = install_fake_orchestration()
```

```python
import histdatacom
from histdatacom.options import Options

options = Options()
options.api_return_type = "polars"
options.formats = {"ascii"}
options.timeframes = {"1-minute-bar-quotes"}
options.pairs = {"eurusd"}
options.start_yearmonth = "2021-04"
options.end_yearmonth = "2021-05"
options.cpu_utilization = "medium"

data = histdatacom(options)
```

```python
print(type(data))
print(data.shape)

if sample_context is not None:
    assert data.shape == (2, 6)
    assert sample_context["request"].api_return_type == "polars"
    assert sample_context["kwargs"] == {
        "start_if_needed": True,
        "wait_for_result": True,
    }
```
